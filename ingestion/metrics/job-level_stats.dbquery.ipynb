```sql
SELECT
    run_id,
    job_id,
    run_name,
    MIN(period_start_time) AS start_time,
    MAX(period_end_time) AS end_time,
    SUM(execution_duration_seconds) AS execution_seconds,
    SUM(setup_duration_seconds) AS setup_seconds,
    SUM(cleanup_duration_seconds) AS cleanup_seconds,
    MAX(result_state) AS result_state
FROM system.lakeflow.job_run_timeline
-- WHERE run_name = 'Yahoo Ingestion'
WHERE period_start_time >= current_date - INTERVAL 1 DAYS
GROUP BY run_id, job_id, run_name
ORDER BY start_time DESC;
```
